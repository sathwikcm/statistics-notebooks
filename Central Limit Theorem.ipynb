```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
import random
random.randint(0,1000)
```

```python
means = []
for i in range(25):
    numbers = []
    for i in range(50):
        numbers.append(random.randint(0,100))
    means.append(np.mean(numbers))
```

```python
sns.distplot(means);
```
